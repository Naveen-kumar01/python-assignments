Q1. What is the concept of an abstract superclass?

Ans - An abstract superclass is also known as abstract base class. In an abstract class contains at least one abstract,
      method or more. An abstract method is the method which does not contain any implementation. But these abstract methods
      must be implemented in the subclass/child class.
      All the abstract methods must be implemented in the child class,failing to do so will raise TypeError.

Q3. Why does a class need to manually call a superclass's __init__ method?

Ans - The main aim of __init__ method is to initialise the variables of the class, hence the manual call for parent class 
      init meyhod will help you to get access of the varibles and funvtions that are present in the base class otherwise 
      you will not be able to access any variable of base class in the inherited class.So if you don't call base class init,
      none of that code would be executed and you would end up with base class that has no variables.

Q4. How can you augment, instead of completely replacing, an inherited method?

Ans - A subclass can either completely override the implementation for an inherited method or the subclass can enhance the
      method by adding functionality to it. Replacing a Superclass's Method Implementation. Sometimes, a subclass will want
      to replace entirely its superclass's implementation of a method. Indeed, many superclasses provide an empty method
      implementation with the expectation that most, if not all, subclasses will completely replace the superclass's
      implementation of the method.
