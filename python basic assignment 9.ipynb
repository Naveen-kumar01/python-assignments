Q6. What are the three “mode” arguments that can be passed to the open() function?

A6 - The modes used for the open() function are : 
     1. r: This mode opens the file in read onlly mode.
     2. rb: This mode opens the file as read-only in binary format and starts reading from the beginning of the file.
     3. r+: This mode opens a file for reading and writing, and the pointer is at the beginning of the file.
