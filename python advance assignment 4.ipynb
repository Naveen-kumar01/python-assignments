```python
Ans - The two operator overloading methods that can be used to support iteration are - 
      1. __iter__ method - This method iterator object and is implicitly called at the beginning of the loops.
      2. __next__ method - This method returns next value and is implicitly called at each loop increment.
```

```python
class Count:
    def __init__(self, l, h):
        self.cur = l
        self.end = h
 
    def __iter__(self):
        return self
 
    def __next__(self):
        if self.cur > self.end:
            raise StopIteration
        else:
            self.cur += 1
            return self.cur - 1
 
 
for num in Count(5, 15):
    print(num)
```

Q2. In what contexts do the two operator overloading methods manage printing?

Ans - If we want the print() function to print the coordinates of the Point object instead of what we got. We can define 
      a __str__() method in our class that controls how the object gets printed. Let's look at how we can achieve this.

```python
class Structure:
    def __init__(self, a=0, b=0):
        self.a = a
        self.b = b

    def __str__(self):
        return "({0}, {1})".format(self.a, self.b)


s = Structure(2, 3)
print(s)
```

Q3. In a class, how do you intercept slice operations?

Ans - In python there is a method called __getitem__ method which is used for iterating list items, array elements, 
      dictionary entries etc. 
      slice is a constructor in Python that creates slice object to represent set of indices that the range(start, stop, step) 
      specifies. __getitem__ method can be implement in a class, and the behavior of slicing can be defined inside it.

```python
sliced ='Cam'.__getitem__(slice(0, len(sliced), 2)) 
print(sliced)
```
